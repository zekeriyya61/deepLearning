```
from imageai.Detection import VideoObjectDetection

#execution_path = "drive/course/deepLearning/ObjectDetection/YOLOv3/"

detector = VideoObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath( "drive/course/deepLearning/ObjectDetection/YOLOv3/yolo.h5")
detector.loadModel()

video_path = detector.detectObjectsFromVideo(input_file_path="drive/course/deepLearning/ObjectDetection/YOLOv3/sampleVideo.mp4",
                                output_file_path="drive/course/deepLearning/ObjectDetection/YOLOv3/sampleVideo_detected_2"
                                , frames_per_second=29, log_progress=True)
print(video_path)

for eachObject in video_path:
    print(eachObject["name"] , " : " , eachObject["percentage_probability"] )
```
