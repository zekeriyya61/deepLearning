```
from imageai.Detection import ObjectDetection
import os

execution_path = os.getcwd()

detector = ObjectDetection()
detector.setModelTypeAsRetinaNet()
detector.setModelPath( "drive/course/deepLearning/ObjectDetection/DetectionFromImage/resnet50_coco_best_v2.0.1.h5")
detector.loadModel()
#custom_objects = detector.CustomObjects(person=True, car=True)
detections = detector.detectCustomObjectsFromImage(input_image="drive/course/deepLearning/ObjectDetection/DetectionFromImage/test7.jpg",
                                                   output_image_path="drive/course/deepLearning/ObjectDetection/DetectionFromImage/test7_detected_2", 
                                                   minimum_percentage_probability=65)

#custom_objects=custom_objects,
for eachObject in detections:
   print(eachObject["name"] + " : " + eachObject["percentage_probability"] )
   print("--------------------------------")
```
